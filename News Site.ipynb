# News site: Get the news articles published on March 11-12.

This notebook gets the following information from a news article publised in Manila Bulletin on March 11-12: `date`, `title`, `full article`, and `author`

### Importing `requests` library

```python
import requests

key = 'e66d4bc5-56e7-4514-a3e4-9022ecf9b7ce'

url = 'https://content.guardianapis.com/search?from-date=2021-03-11&to-date=2021-03-12&page-size=5&show-fields=body&api-key='+key
r1 = requests.get(url)
json_news = r1.json()
```

```python
#Finding how many Pages
pages = json_news['response']['pages']
pages
```

```python
url_article = 'https://www.theguardian.com/sport/2021/mar/12/rory-mcilroy-chasing-bryson-dechambeaus-power-has-cost-me'
article=requests.get(url_article)

from bs4 import BeautifulSoup
import re, datetime

soup = BeautifulSoup(article.content, 'html.parser')

# Getting the Title
art_title = json_news['response']['results'][0]['webTitle']
print(f'{art_title} \n')

# Getting the Author
author = soup.find_all("a", {"rel": "author"})
art_authors = author[0].text
for i in range(1,len(author)):
    art_authors = art_authors+' and '+author[i].text
print(f'{art_authors}')

# Getting the Date
date = json_news['response']['results'][0]['webPublicationDate']
match = re.search('\d{4}-\d{2}-\d{2}',date)
art_date = datetime.datetime.strptime(match.group(), '%Y-%m-%d').date()
print(f'{art_date} \n')

#Getting the content
fullcontent = json_news['response']['results'][0]['fields']['body']
print(f'{fullcontent}')
```
